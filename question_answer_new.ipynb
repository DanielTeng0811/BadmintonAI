```python
import platform
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# 字體設定
system = platform.system()
if system == 'Darwin':
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'PingFang TC', 'Heiti TC', 'sans-serif']
elif system == 'Windows':
    plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'SimHei', 'sans-serif']
else:  # Linux (Streamlit Cloud)
    plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei', 'DejaVu Sans', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False

# 1. 篩選資料
match_id = 1.0
player_name = 'CHOU Tien Chen'

match_data = df[df['match_id'] == match_id].copy()
player_smashes = match_data[(match_data['player'] == player_name) & (match_data['type'] == '殺球')].copy()

# 2. 計算殺球頻率
smash_frequency = player_smashes.groupby('set').size()
total_shots = match_data[match_data['player'] == player_name].groupby('set').size()
smash_ratio = (smash_frequency / total_shots) * 100

# 3. 計算殺球得分率
def is_smash_winning(row):
    rally_group = match_data[(match_data['set'] == row['set']) & (match_data['rally'] == row['rally'])]
    if rally_group.empty:
        return False
    last_shot = rally_group['getpoint_player'].iloc[-1]
    if pd.isna(last_shot):
        return False
    return last_shot == player_name
    

player_smashes['is_winning_smash'] = player_smashes.apply(is_smash_winning, axis=1)
smash_scores = player_smashes.groupby('set')['is_winning_smash'].sum()

smash_counts = player_smashes.groupby('set').size()

# 避免除以零的情況
smash_success_rate = (smash_scores / smash_counts) * 100
smash_success_rate = smash_success_rate.fillna(0) # 將 NaN 設為 0

# 將 Series 轉換為 DataFrame
smash_frequency_df = smash_frequency.rename('smash_frequency').to_frame()
smash_ratio_df = smash_ratio.rename('smash_ratio').to_frame()
smash_success_rate_df = smash_success_rate.rename('smash_success_rate').to_frame()

# 合併 DataFrame
summary_df = pd.concat([smash_frequency_df, smash_ratio_df, smash_success_rate_df], axis=1).reset_index()

# 4. 統計檢定
# 计算局數與殺球頻率的相關係數
if len(summary_df['set']) > 1:
    correlation_frequency, p_value_frequency = pearsonr(summary_df['set'], summary_df['smash_ratio'])

    # 计算局數與殺球得分率的相關係數
    correlation_success_rate, p_value_success_rate = pearsonr(summary_df['set'], summary_df['smash_success_rate'])

    print("\n統計檢定結果:")
    print(f"局數與殺球頻率的相關係數: {correlation_frequency:.2f}, p-value: {p_value_frequency:.3f}")
    print(f"局數與殺球得分率的相關係數: {correlation_success_rate:.2f}, p-value: {p_value_success_rate:.3f}")
else:
    print("局數太少，無法進行相關性分析。")
    correlation_frequency, p_value_frequency, correlation_success_rate, p_value_success_rate = None, None, None, None

# 5. 繪製圖表
fig, axes = plt.subplots(2, 1, figsize=(12, 14))

# 殺球頻率
axes[0].plot(summary_df['set'], summary_df['smash_ratio'], marker='o', linestyle='-', color='#FF6B6B')
axes[0].set_title(f'第 {int(match_id)} 場比賽 - {player_name} 每局殺球頻率', fontsize=16, fontweight='bold')
axes[0].set_xlabel('局數 (Set)', fontsize=12)
axes[0].set_ylabel('殺球頻率佔總擊球次數 (%)', fontsize=12)
axes[0].set_xticks(summary_df['set'])
axes[0].grid(True, alpha=0.3, linestyle='--')

# 殺球得分率
axes[1].plot(summary_df['set'], summary_df['smash_success_rate'], marker='o', linestyle='-', color='#4ECDC4')
axes[1].set_title(f'第 {int(match_id)} 場比賽 - {player_name} 每局殺球得分率', fontsize=16, fontweight='bold')
axes[1].set_xlabel('局數 (Set)', fontsize=12)
axes[1].set_ylabel('殺球得分率 (%)', fontsize=12)
axes[1].set_xticks(summary_df['set'])
axes[1].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()

print(summary_df)

# 6. 結論
print("\n結論:")
if correlation_frequency is not None and correlation_success_rate is not None:
    print("根據統計檢定結果，")
    if correlation_frequency < 0 and p_value_frequency < 0.05:
        print("殺球頻率隨著局數增加有顯著下降的趨勢。")
    else:
        print("殺球頻率隨著局數增加沒有顯著下降的趨勢。")

    if correlation_success_rate < 0 and p_value_success_rate < 0.05:
        print("殺球得分率隨著局數增加有顯著下降的趨勢。")
    else:
        print("殺球得分率隨著局數增加沒有顯著下降的趨勢。")
else:
    print("由於局數太少，無法判斷殺球頻率和得分率是否隨著局數增加而顯著下降。")
```
